```python
# 4. Count the Number of Lines

# Write a program to count how many lines are present in notes.txt.
with open("notes.text","r") as f:
    count=0
    for line in f:
        count+=1
    print(count)
```

```python
# Write a program to count the total number of words present in notes.txt.
with open("notes.text","r") as f:
    count=0
    for line in f:
        a=line.split()
        count+=len(a)
    print(count)
```

```python
# 8. Find the Longest Word
# # Write a program to find and print the longest word present in notes.txt.
with open("notes.text","r") as f:
    print(f.read())
    f.seek(0)  
    words=f.read().split()
longest=words[0]
for ele in range(len(words)):
    if len(words[ele])>len(longest):
        longest=words[ele]
print("Longest word is : ",longest,"\n")
print(f"{longest} has length {len(longest)}")
    
    
```

```python
# 9. Count Occurrences of a Specific Word

# Write a program that asks the user to enter a word and counts how many times that word appears in notes.txt.

# Example:

# Enter word: Python
# Python appears 1 time(s).

# (Ignore case differences, e.g., python and Python should be treated as the same word.)
word=input("Enter any word: ")
count=0
with open("notes.text","r") as f:
    a=f.read().split()
    for ele in a:
        if word.lower() == ele:
            count+=1
print(f"{word} appear {count} times")
```

```python
# 10. Replace a Word in a File

# Write a program that replaces all occurrences of the word "Python" with "Java" in notes.txt and saves the changes back to the same file.

# Before:

# Python is easy to learn.

# After:

# Java is easy to learn.
with open("notes.text","r") as f:
    print(f.read().lower().replace("python","java"))
```

```python
# 11. Create a Student Record File

# Create a file named students.txt. Continuously ask the user to enter:

# Student Name
# Marks

# Store each record in the format:

# Aman,85
# Riya,92
# Karan,78

# Stop taking input when the user enters "exit" as the name. Then:

# Display all records.
# Find the student with the highest marks.
with open("student.txt","w+") as f:
   
    while True:
        name=input("Enter Student name or enter exit to stop")
        if name.lower()=="exit":
            break
        marks=int(input("Enter student\'s marks"))
        f.write(f"{name},{marks}\n")
    f.seek(0)
    print(f.read())
    
    
```

```python
# finding the student with highest marks
highest = -1
topper = ""

with open("student.txt", "r") as f:
    for line in f:
        print(type(line))
        name, marks = line.strip().split(",")
        marks = int(marks)

        if marks > highest:
            highest = marks
            topper = name

print(f"Topper: {topper}")
print(f"Highest marks are {highest}")
            
```
